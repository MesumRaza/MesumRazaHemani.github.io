```python
import youtube_dl

ydl = youtube_dl.YoutubeDL({'skip_download':True})

with ydl:
    result = ydl.extract_info(
        'https://www.youtube.com/playlist?list=PLQhvMBdV3VIHJX8d1wqr7Ig_py56V6uAQ',
        #'https://www.youtube.com/playlist?list=UU5kd-4D7J4zbhqjFMWJ3pgQ', #REHAN ALLAHWALA PLAYLIST 6689 VIDEOS
        download=False # We just want to extract the info
    )
```

```python
selected_attributes=['title','upload_date','duration','view_count','average_rating','webpage_url','categories','like_count','dislike_count','resolution']
list_videos=[{key: value for key, value in x.items() if key in selected_attributes} for x in result['entries']]
```

```python
import pandas as pd
df = pd.DataFrame(list_videos,columns=selected_attributes)
```
